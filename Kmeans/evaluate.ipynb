```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
means = [[-1,-1],[1,1]]
covs = [[0.25,0],[0,0.25]]
n = 100
np.random.seed(42)

X = np.vstack([np.random.multivariate_normal(i, covs, n) for i in means])
y = np.repeat([0,1],n)
plt.scatter(X[:,0], X[:,1], c=y, cmap='jet')
plt.axis('equal')
plt.show()
```

```python
from kmeans import KMeans

k = KMeans(2, 200)
history = k.fit(X)
```

```python
plt.scatter([x[0] for x in k.clusters], [x[1] for x in k.clusters], c='black', marker='x')
plt.scatter(X[:,0], X[:,1], alpha=.1)
plt.show()
```

```python
plt.figure(figsize=(16,4))
for i in range(len(history.keys())):
  plt.subplot(2,4, i+1)
  plt.scatter(X[:,0], X[:,1], alpha=.1)
  plt.scatter([x[0] for x in history[str(i)]], [x[1] for x in history[str(i)]], c='black', marker='x')
  plt.xticks([])
  plt.yticks([])
  plt.xlabel(f'{i}')
plt.show()
```

```python
np.random.shuffle(X)

new_data = X[2:5]
new_data = np.append(new_data, np.array([-10, -5])).reshape((4,2))
labels = k.predict(new_data)

plt.scatter([x[0] for x in k.clusters], [x[1] for x in k.clusters], c='black', marker='x')
plt.scatter(new_data[0,0], new_data[0,1], c='red', label=labels[0])
plt.scatter(new_data[1,0], new_data[1,1], label=labels[1])
plt.scatter(new_data[2,0], new_data[2,1], label=labels[2])
plt.scatter(new_data[3,0], new_data[3,1], label=labels[3])
plt.scatter(X[:,0], X[:,1], alpha=.01, c='blue')
plt.legend()
plt.show()
```
