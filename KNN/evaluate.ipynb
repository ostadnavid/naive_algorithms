```python
import numpy as np
import matplotlib.pyplot as plt
```

## classification

```python
means = [[-1,-1],[1,1]]
covs = [[0.25,0],[0,0.25]]
n = 100
np.random.seed(1223)

X = np.vstack([np.random.multivariate_normal(i, covs, n) for i in means])
y = np.repeat([0,1],n)
plt.scatter(X[:,0], X[:,1], c=y, cmap='jet')
plt.axis('equal')
plt.show()
```

```python
from knn import KNN

knn_predictor = KNN('classification', k=5)

knn_predictor.fit(X, y)

knn_predictor
```

```python
n = 10
rand_int1 = np.random.randint(0, (len(X)//2)-n//2)
rand_int2 = np.random.randint((len(X)//2)-n//2, len(X)-n)

X_test = np.concatenate((X[rand_int1:rand_int1+n//2], X[rand_int2: rand_int2+n//2]))
y_test = np.concatenate((y[rand_int1:rand_int1+n//2], y[rand_int2: rand_int2+n//2]))

X_test, y_test
```

```python
preds = knn_predictor(X_test)

plt.scatter(X[:,0], X[:,1], c=y, alpha=.05, cmap='jet')
plt.scatter(X_test[:,0], X_test[:,1], c=preds, cmap='jet', marker='x')
plt.axis('equal')
plt.show()
```
